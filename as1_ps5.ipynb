Overview
PostgreSQL database that stores user information. Uses Drizzle ORM which is a database tool for TypeScript.

What it does
- Stores user accounts with username and password
- Tracks how many times each user analyzes resumes
- Manages premium/paid tier users
- Keeps track of when users joined and updated their info
- Uses Clerk for user authentication

That's it. Simple database for user management.
