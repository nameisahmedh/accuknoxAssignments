stroke_ml.py - Machine Learning Module

GitHub: https://github.com/nameisahmedh/strokeprediction/blob/main/stroke_ml.py

Overview
Python module that predicts stroke risk. It cleans data, trains machine learning models, and makes predictions.

What it does
- Prepares data by filling missing values, encoding text, scaling numbers
- Trains 7 different machine learning models (RandomForest, XGBoost, CatBoost, SVM, KNN, LogisticRegression, NaiveBayes)
- Compares which model works best
- Makes predictions on new data
- Saves and loads trained models

Best model: XGBoost with 95% accuracy
