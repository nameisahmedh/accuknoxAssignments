### Adding csv into database

```python
import sqlite3
import pandas as pd

db_name = 'user_data.db'
table_name = 'users'
csv_file = 'sample.csv'

conn = sqlite3.connect(db_name)
cursor = conn.cursor()

cursor.execute(f'''
    CREATE TABLE IF NOT EXISTS {table_name} (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        email TEXT NOT NULL UNIQUE
    );
''')
conn.commit()

df = pd.read_csv(csv_file)
# Changed if_exists='append' to if_exists='replace' to avoid UNIQUE constraint errors on re-runs
df.to_sql(table_name, conn, if_exists='replace', index=False)

print("Data imported and verified:")
query_df = pd.read_sql_query(f"SELECT * FROM {table_name};", conn)
print(query_df)

conn.close()
```
